## <font color=blue>Thu thập dữ liệu</font>

```python
import requests
import json
```

Hiện giờ, GitHub cho phép dùng API theo một số cách sau:
- Không chứng thực (nghĩa là cứ thế mà dùng thôi, không cần thủ tục lôi thôi). Cách này bị giới hạn về số lần request / đơn vị thời gian; chẳng hạn, với API tìm kiếm thì bị giới hạn tối đa là 10 request / phút.
- Có chứng thực bằng (i) username và password của tài khoản GitHub, hoặc (ii) token phát sinh ra từ tài khoản GitHub. Cách này thì được nhiều lần request hơn; chẳng hạn, với API tìm kiếm thì là 30 request / phút.

Trong bài này, bạn sẽ dùng API của GitHub theo cách có chứng thực bằng token. Đầu tiên, bạn cần có một account trên GitHub. Sau đó, bạn vào [đây](https://github.com/settings/tokens) để phát sinh ra một token; ở phần `Select scopes`, bạn không cần phải check gì cả (trong bài này, không cần phải dùng đến những quyền lực đó). Một cách sử dụng token là sử dụng giống như password. (Sau khi có điểm bài này, bạn có thể thu hồi token lại.)

```python
# Dùng Github API để tìm kiếm các thùng chứa với từ khóa "data science",
# và chỉnh để được 100 kết quả / trang (mặc định là 30 kết quả / trang)
user_name = '1512473' # TODO
token = 'd582e01e70aac57f361e662924cb1dc05bec74ee' # TODO
r = requests.get('https://api.github.com/search/repositories?q=data%20science&per_page=100', 
                 auth=(user_name, token))
r.ok # Nếu không có vấn đề gì thì kết quả sẽ là True
```

```python
# Chuyển từ chuỗi JSON về cấu trúc dữ liệu của python
py_data = json.loads(r.text)
py_data
```

```python
py_data.keys()
```

`total_count` là tổng số kết quả tìm kiếm được (nhưng như đã nói ở trên, GitHub API chỉ cho lấy tối đa là 1000 kết quả thôi, có thể có cách để lấy hơn 1000 nhưng tạm thời ta không bàn tới), `items` là list chứa các kết quả của trang đầu tiên (trang 1).

```python
py_data['total_count']
```

```python
len(py_data['items'])
```

```python
# Xem headers của biến `r` để biết một số thông tin như:
#    X-RateLimit-Limit: số request tối đa / phút
#    X-RateLimit-Remaining: số request còn lại
#    Link: chứa link tới trang kết quả kế và link tới trang kết quả cuối;
#          ta có thể dựa vào đây để lấy hết các kết quả

for key, value in r.headers.items():
    print('{}: {}'.format(key, value))
```

```python
# TODO: Lần lượt duyệt các trang và thu thập hết 1000 kết quả
results = [] # List dùng để chứa 1000 kết quả

for i in range (1,11):
    rq_str = 'https://api.github.com/search/repositories?q=data%20science&per_page=100&page=' + str(i)
    r_i = requests.get(rq_str, auth=(user_name, token))
    py_data = json.loads(r_i.text)
    results += py_data['items']
    
```

```python
# Nên lưu dữ liệu thu thập được xuống file trước khi làm tiếp
f = open('ds_repositories.json', 'w')
json.dump(results, f, indent=2) # dump này không có s
f.close()
```
